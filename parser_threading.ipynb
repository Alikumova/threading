```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import threading
import time
import re
eM = []
def findEmail(domain):
    try:
        page_3_dom = urlopen(domain)
    except Exception:
            print('FE: open error, dom: ' + domain)
            return
    try:
        soup_3_dom = BeautifulSoup(page_3_dom, 'html.parser')
    except Exception:
            print('FE: parseBS error, dom: ' + domain)
            return
    try:
        email = re.findall(r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,4}',str(soup_3_dom))
    except Exception:
            print('FE: parseBS error, dom: ' + domain)
            return
    email = list(set(email))
    if 'Rating@Mail.ru' in email:
        email.remove('Rating@Mail.ru')
    for element in email:
            eM.append(element)

def startFinder(domain):
    #Отчитываемся о том, что запустили поток
    print("thread " + domain + " start \n")
    try:
        page_3 = urlopen('http://isthissitebuiltwithdrupal.com/' + domain, timeout=10)
        soup_3 = BeautifulSoup(page_3, 'html.parser')
        result = soup_3.findAll(attrs={"class":"answer"})
    except Exception:
        print(domain, 'o/p error')
        return 
    
    if str(result).find('yes') > 0:
        print(domain + ' yes ')       
        findEmail(domain)
    elif str(result).find('no') > 0:
        print(domain + ' no')
    
file = open('domains_list.txt','r')
domains = []
for string in file:
    domains.append(string.replace('\n',''))
file.close()
t1 = time.time()
events = []
for domain in domains:
    events.append(threading.Thread(target=startFinder,args=(domain,)))
    #Запускаем поток
    events[-1].start()
#Завершаем потоки
for e in events:
    e.join()
print(time.time() - t1)
file = open('email_list.txt','w')
for element in eM:
    file.write(element + '\n')
file.close()
```
